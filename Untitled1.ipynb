```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import xarray as xr
import random
from scipy.stats import norm
from scipy.integrate import trapz

%matplotlib inline
```

```python
x = np.arange(0, 2, 0.001) # entire range of x, both in and out of spec
x_aice = np.arange(0, 1, 0.001)
mod_points = random.sample(list(x), 10)
aice_points = random.sample(list(x_aice), 10)
obs = random.choice(x)
y = np.zeros(len(mod_points))
y_up = y + 0.1

y_mod= norm.pdf(x,np.mean(mod_points),0.5)
y_obs = norm.pdf(x,obs,0.1)
y_pos = y_mod*y_obs
```

```python
norm_obs = trapz(y_obs, dx = 0.001)
norm_mod = trapz(y_mod, dx = 0.001)
norm_pos = trapz(y_pos, dx = 0.001)
```

```python
observations = y_obs/norm_obs
model = y_mod/norm_mod
posterior = y_pos/norm_pos

kalman = 0.5/0.8
difference = (np.zeros(len(mod_points))+obs) - mod_points
increments = kalman*(difference)
update = mod_points + increments
```

```python
fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('seaborn-white')
ax.set_xlim([-0.03,1])
ax.set_ylim([-0.03,2])
ax.set_xlabel('sea ice freeboard (m)', fontsize=14)
ax.set_yticklabels([])

ax.plot(mod_points, y, '*', c = 'maroon', markersize=12)
ax.plot(x, y_mod, c = 'maroon', linewidth=3)

ax.plot(obs, 0, 's', c = 'seagreen', markersize=10)
ax.plot(x, y_obs, c = 'seagreen', linewidth=3)

ax.plot(x, posterior, c = 'royalblue', linewidth=3)
ax.plot(update, y_up, '^', c = 'royalblue', markersize=10)
```

```python
x_free = np.arange(0, 5, 0.001) # entire range of x, both in and out of spec
x_aice = np.arange(0, 1, 0.001)

EE = np.arange(0, 1, 0.001)
error = random.sample(list(EE), 15)

free_points=np.zeros(len(error))
aice_points=np.zeros(len(error))

omod_points = sorted(random.sample(list(x_free), 15))
oobs_points = random.sample(list(x_aice), 15)

for i in range(0, len(error)):
    free_points[i] = error[i]*omod_points[i]
    aice_points[i] = error[i]*oobs_points[i]
```

```python
m,b = np.polyfit(free_points[:-1], aice_points[:-1],1)
line = m*free_points[:-1] + b
```

```python
fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('seaborn-white')

ax.plot(free_points[:-1],aice_points[:-1], 'H', c = 'maroon', markersize=12)
# ax.plot(free_points[:-1], line, c = 'maroon', linewidth=3)
ax.set_ylim([-0.05,1])
ax.set_xlim([-0.05,4.1])
ax.set_xlabel('sea ice freeboard (m)', fontsize=16)
ax.set_ylabel('sea ice concentration', fontsize=16)
```

```python
import numpy as np
import numpy.ma as ma
import xarray as xr
import scipy as sc
from matplotlib import pyplot as plt
%matplotlib inline
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cmocean as cmo
import os.path
import nc_time_axis
```

```python
## Import unperturbed SOM runs from MW and CMB directories

ice={} #empty dictionary for file paths
ices={} #empty dictionary for datasets

path = '/glade/scratch/bitz/free10d/run/'


for x in range(1,10):
    for xx in range(1,10):
        ice["file000{0}_0{1}".format(x,xx)]= path + 'free10d.cice_000{0}.h.2000-01-0{1}.nc'.format(x,xx)
        ices["FREE_000{0}_0{1}".format(x,xx)] = xr.open_dataset(ice["file000{0}_0{1}".format(x,xx)])

        
    for xx in range(10,32):
        ice["file000{0}_{01}".format(x,xx)]= path + 'free10d.cice_000{0}.h.2000-01-{01}.nc'.format(x,xx)
        ices["FREE_000{0}_{01}".format(x,xx)] = xr.open_dataset(ice["file000{0}_{01}".format(x,xx)])
```
